# TRABAJO CON FICHEROS

| Character | Meaning                                                |
|-----------|--------------------------------------------------------|
| ‘r’       | open file for reading (default)                        |
| ‘w’       | open for writing. If file exists, replace its contents |
| ‘a’       | open for writing. If file exists, appends to end       |
| ‘b’       | binary mode                                            |
| ‘t’       | text mode (default)                                    |
| ’+’       | reading and writing                                    |

#### READING

```python
with open("joflorid.txt", 'rb') as fd:
	print(type(fd))
	lines = fd.readlines()
	print(type(lines))
for line in lines:
	print(line)
```

```python

with open("joflorid.txt", 'r') as fd:
	while True:
		data = fd.read(40) # Con cada iteracion del while, lee 20 bytes
		#print(type(data))
		if not data:
			break
		print(data)
```

#### WRITTING

Warning: Si se utiliza el caracter 'w' para escribir en un archivo, el contenido del mismo se reemplaza por la nueva escritura. Si lo que se quiere es añadir contenido al ya existente, el caracter a usarse seria 'a'

##### Escribir en un fichero que no existe (se crea)

```python
with open("joflorid_2.txt", 'w') as fd:
	fd.write("Esta linea ha sido escrita desde Jupyter Notebook")
```

```python
str_list = ["linea1\n", "linea2\n", "linea3\n", "linea4"]
with open("joflorid_3.txt", 'w') as fd:
	fd.writelines(str_list)
```

Añadir escritura a un fichero que ya existe sin destruir la informacion que ya contiene

```python
with open("joflorid.txt", 'a') as fd:
	fd.write("Esta si es la ultima linea escrita desde Jupyter Notebook...")
```
