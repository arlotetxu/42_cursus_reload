# STRINGS

Para conocer los métodos que aplican a los strings, se puede ejecutar __dir("")__

```python
name = 'Jose'
age = 46
```

```python
print("My name is: %s. I'm %i years old." % (name, age))
```

```python
print("My name is: {}. I'm {} years old.".format(name, age))
```

```python
print("My name is: {first_name}. I'm {age} years old.".format(first_name=name, age=age))
```

```python
print("My name is: {first_name}. I'm {age} years old.".format(**{'first_name': name, 'age': age}))
```

Los ** hacen que a la funcion 'print' se pasen los valores del diccionario como argumentos.

*__La mas común:__*

```python
print(f"My name is: {name.upper()}. I'm {age} years old.")
```

```python
simple_dict = {'name': 'Jose', 'age': 46}
print(f"My name is {simple_dict['name']}. I'm {simple_dict['age']}.")
```

### ALINEACIÓN

```python
name = 'Jose'
print(f"{name:<20}") #Alinea el nombre a la izquierda en un array de 20 caracteres
```

```python
print(f"{name:>20}") #Alinea el nombre a la derecha en un array de 20 caracteres
```

```python
print(f"{name:^20}") #centra el nombre en un array de 20 caracteres
```

### CONCATENAR

```python
string_1 = "Hello "
string_2 = "World!."
string_3 = string_1 + string_2
print(f"{string_3}")
```

```python
'--'.join([string_1, string_2]) # lo que ponemos entre comillas al inicio es lo que se usa para unir las cadenas
```

### SPLITTING

```python
string_4 = "Esto es una cadena de 7 palabras"
string_4.split() # Por defecto, se usa el espacio como caracter delimitador de palabras
```

```python
string_5 = "Esto;es;una;cadena;de;7;palabras" # Util para procesar archivos CSV o similar
string_5.split(';')
```

### SLICING

```python
string_5[1:5] # Desde el índice 1 hasta el indice 4. El numero al final es el índice limite al que nunca se llega
```

```python
string_5[:5]
```

```python
string_5[:-4]
```

```python
string_5[-4:] # Extrae los 4 últimos caracteres
```
