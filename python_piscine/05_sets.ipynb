# SETS

Colecci칩n desordenada de DISTINTOS objetos. Debido a que los elementos se encuentran desordenados, no tienen un registro por indices. Hay 2 tipo se sets:
- set --> mutable
- frozenset --> inmutable

#### CREACION
(se crean como una serie de objetos separados por comas y dentro de llaves (curly brackets)

```python
my_set = {"a", "b", "c", "c"}
print(my_set) # Observa que se eliminan los elementos repetidos y estan desordenados
```

```python
my_list = [1, 2, 22, 2]
my_set = set(my_list)
print(my_set)
```

### ACCEDIENDO A SUS MIEMBROS

Un set no permite usar slicing ([:2]). Para acceder a sus miembros, hay que iterar sobre los mismos con un loop

```python
my_set[0] # Provocando el error
```

```python
my_set[:1] # Provocando el error
```

```python
for item in my_set:
    print(item)
```

### METODOS

```python
[x for x in dir(my_set) if not x.startswith("__")]
```

##### __.add()__
(A침ade un item a un set)

```python
my_set.add(33)
print(my_set)
```

##### __.update()__
(A침ade multiples items a un set. Es necesario pasar un iterable)

```python
my_set.update((44, 55))
print(my_set)
```

##### __.remove__
(Elimina un item especifico de un set)

```python
my_set.remove(55)
print(my_set)
```

```python
my_set.remove(888) # Provocando el error al no tener el item 888 en el set
print(my_set)
```

##### __.discard()__
(Funciona de una manera similar a .remove pero no muestra error en caso de que el item especificado no se encuentre en el set)

```python
my_set.discard(888)
print(my_set)
```

##### __.pop()___
(Elimina y retorna un item *ARBITRARIO* de un set)

```python
print(f"{my_set.pop()}")
print(my_set)
```

##### __.clear()__
(Vacia un set)

```python
print(my_set)
my_set.clear()
print(my_set)
```

##### __del__
(Elimina un set)

```python
del(my_set)
print(my_set) # Provocando el error
```

##### __.union()__
(Combina 2 sets y retornando la combinaci칩n en un nuevo set)

```python
my_set1 = {"a", "b", "c", "d"}
my_set2 = {"z", "x", "y"}
my_set3 = my_set1.union(my_set2)
print(my_set3)
```

##### __.intersection()__
(Retorna un nuevo set con los elementos comunes de 2 sets)

```python
my_set1 = {"a", "b", "c", "d"}
my_set2 = {"z", "a", "y"}
my_set3 = my_set1.intersection(my_set2)
print(my_set3)
```

##### __.difference()__
(Retorna un nuevo set con los elementos diferentes de 2 sets)

```python
my_set1 = {"a", "b", "c", "d"}
my_set2 = {"z", "a", "y"}
my_set3 = my_set1.difference(my_set2)
print(my_set3)
```

Para eliminar duplicados en una lista...

```python
my_list = [1, 2, 3, 4, 2, 4]
my_set = set(my_list)
print(my_set)
my_list = list(my_set)
print(my_list)
```
