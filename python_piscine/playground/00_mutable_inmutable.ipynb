# MUTABLE vs INMUTABLE

Un objeto mutable es aquel que se puede modificar sin destruirlo. POr el contrario, uno inmutable, para modificarlo se destruye y se crea uno nuevo.

INMUTABLES: numeros (int, float,...), strings y tuplas

MUTABLES: listas, diccionarios y set o conjuntos

### PROBANDOLO

```python
# CON NUMEROS
a = 5
print(id(a))
a = 7
print(id(a)) # El ID tras la modificación es distinto del primero por lo que se ha creado un nuevo objeto
```

```python
# CON STRINGS
str = "Esto es una cadena"
print(id(str))
str += ". Aqui la modificación"
print(id(str))
```

```python
# Con tuplas
my_tupla = (2, 3, 'a')
print(type(my_tupla))
print(id(my_tupla))
my_tupla = tuple([2, 3, 'a', 4])
print(id(my_tupla))
```

```python
# Con listas
my_list = [1, 2, 3]
print(id(my_list))
my_list.append(4)
print(id(my_list))
```

```python
# Con diccionarios
my_dict = {'first_name': 'Jose', 'Age': 46}
print(id(my_dict))
my_dict['last_name'] = 'Florido'
print(id(my_dict))
print(my_dict)
```

```python
# Con sets
my_set = {"a", "b", "c"}
print(id(my_set))
my_set.add("d")
print(id(my_set))
```
