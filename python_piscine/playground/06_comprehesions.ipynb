# PYHTON COMPREHENSIONS

Se muestran diferentes maneras de crear iterables. Normamente se usan las comprehensions cuando hay que generar un nuevo iterable y hay que realizar alguna operación sobre los elementos. A continuación algunos ejemplos.

### SALIDA A LISTA

```python
my_list = [1, 2, 3]
my_new_list = [x for x in my_list]
my_new_list
```

```python
my_new_list = [x * 2 for x in my_list]
my_new_list
```

Con condicionales

```python
my_new_list = [x for x in range(0, 10) if x % 2 != 0] # numeros impares entre 0 y 9
my_new_list
```

Con diccionarios

```python
my_dict = {1: 'dog', 2: 'cat', 3: 'python', 4: 'bird'}
[(key, value) for key, value in my_dict.items()]
```

```python
[(key, value) for key in my_dict for value in my_dict.values() if my_dict[key] == value]
```

Anidaciones

```python
new_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
[[item * 2 for item in element]for element in new_list]
```

### SALIDA A DICCIONARIO

```python
my_list = ["dog", "cat", "snake"]
{key: value for key, value in enumerate(my_list)}
```

```python
my_list = ["dog", "cat", "snake"]
{key: value
 for value in my_list
 for key in range(0, len(my_list)) }
```

### SALIDA A SET

```python
my_list = list("aaabbbcccccddddd")
print(my_list)
sorted({item for item in my_list})
```
