# LISTAS

Las listas son un tipo de dato fundamental en python. Son una secuencia de elementos MUTABLES (se pueden modificar). Es posible tener listas anidadas unas dentro de otras. Incluso pueden contener otros tipos de datos como diccionarios u otros objetos. Se definen con [ ].

### CREACION DE LISTAS

```python
my_list = [] # Esto genera una lista vacia
my_list = list() # Esto tb crea una lista vacia
print(f"{my_list}")
```

```python
my_list = [1, 2, 3]
print(f"{my_list}")
print(f"{my_list[0]}")
```

```python
my_list = list('445 abc') # El método list solo acepta un argumento
print(f"{my_list}")

my_tupla = (4, 5, 6)
my_list = list(my_tupla)
print(f"{my_list}")
```

### METODOS

```python
[x for x in dir(my_list) if not x.startswith("__")] # Se muestran los métodos normales, no los especiales o magicos (aquellos que comienzan con __)
```

##### __.count()__
(Cuenta el numero de ocurrencias el argumento en la lista)

```python
my_list = list('abcc')
my_list.count('a') # Cuenta el numero de veces que que tenemos 'a' en la lista my_list
```

```python
my_list.count('c')
```

##### __.index()__
(Devuelve el indice de la primera ocurrencia del argumento en la lista)

```python
my_list.index('c')
```

##### __.reverse()__
(Da la vuelta a la lista. Devuelve None por lo que no se puede asignar a otra variable)

```python
print(f"{my_list}")
my_list.reverse()
print(f"{my_list}")
```

##### __.append()__
(Añade un elemento al final de una lista)

```python
my_list.append(1)
print(f"{my_list}")
len(my_list) # Devuelve el numero de elementos que tiene la lista
```

##### __.insert()__
(Inserta un elemento en la posicion indicada en los argumentos)

```python
print(f"{my_list}")
my_list.insert(0, "Hello")
print(f"{my_list}")
```

### __.extend()__
(Añade un objeto iterable a una lista. Esto es mas eficiente que crear una tercera lista y unir las 2 en la tercera)

```python
my_list_2 = [22, 33, 44]
my_list.extend(my_list_2)
print(f"{my_list}")
```

```python
my_list_3 = ['u', 't', 'h']
my_list += my_list_3 # Similar a extend()
print(f"{my_list}")
```

```python
print(f"{my_list[0]}")
print(f"{my_list[2]}")
print(f"{my_list[-1]}") # El ultimo elemento de la lista
print(f"{my_list[12]}") # Error forzado
```

##### __.clear()__
(Limpia el contenido de una lista)

```python
my_list.clear()
print(f"{my_list}")
```

##### __.pop()__
(Extrae un elemento de la lista acorde al indice pasado como argumento y lo retorna. Por defecto extrae el ultimo elemento)

```python
my_list = list('abbcde')
print(f"{my_list}")
my_list.pop()
print(f"{my_list}")
popped = my_list.pop(0) #Eliminamos el primer valor (indice 0) y lo retornamos a la variable popped
print(f"{my_list}")
print(f"popped: {popped}")
```

##### __.remove()__
(Elimina la primera ocurrencia de un valor especificado como parametro de la lista)

```python
print(f"{my_list}")
my_list.remove('b') # Eliminamos 'b' de la lista
print(f"{my_list}")
```

##### __del__
(Elimina un valor de la lista pasado por su indice)

```python
del my_list[0]
print(f"{my_list}")
del my_list[5] # Error forzado
```

##### __.sort()__
(Ordena la lista. Modifica en la propia lista el orden)

```python
my_list = list('bvtazt')
my_list.sort()
print(f"{my_list}")
my_list.sort(reverse=True)
print(f"{my_list}")
my_list_2 = my_list.sort()
print(f"{my_list_2}") # El orden se realiza en la propia lista por lo que no se retorna nada.
```

##### __sorted(list)__
(Retorna una nueva lista ordenada, asignable a variable)

```python
my_list = list('bvtazt')
my_list_3 = sorted(my_list);
print(f"{my_list}")
print(f"{my_list_3}")
my_list_3 = sorted(my_list, reverse=True)
print(f"{my_list_3}")
```

##### __.copy()__
(Crea una copia de una lista. Ojo que es una Copia superficial (shallow copy), lo que quiere decir que lo que se copia es el contenedor y que los valores tanto de la lista original como de la copia apuntan al mismo sitio. Esto se traduce en que si cambiamo algun valor de la lista original, se cambiará tambien en la copia. 

DISCLAIMER: Creo que esto no se puede comprobar en Jupyter Notebooks)

```python
my_list_copied = my_list.copy()
print(f"{my_list_copied}")
```

```python
my_list_copied.clear()
my_list_copied = my_list[:] # Alternativa a usar la funcion .copy() pero con similares resultados (swallow copy)
print(f"{my_list_copied}")
```

```python
my_list_copied.clear()
my_list_copied = list(my_list) # Alternativa a usar la funcion .copy() pero con similares resultados (swallow copy)
print(f"{my_list_copied}")
```
