Creating a function to visualize data:

```python
def visualize(theta, X, y,):
  plt.scatter(X, y)
  plt.plot(X, theta[0] + theta[1] * X, color='red')

  plt.show()
```

Define a function to get the prediction. **Theta** is an array that contains in its **0 position the intercept or bias** (value where the line crosses the y axis) and **gradient or slope in its 1 position**.

```python
def predict(X, theta):
  #m, c = np.polyfit(X, y, 1)
  #print(m, c)
  y_pred = X * theta[1] + theta[0]
  return y_pred
```

Define the function to train our model. Alpha is the learning rate. It controls how big of a step the algorithm takes during each iteration to update the theta values.

```python
def fit(X, y, theta, alpha, num_iters):
    """
    This loop runs the gradient descent process for the specified number of
    iterations. In each iteration:

    predictions = theta[0] + theta[1] * X: The model makes a prediction y_hat
    using the current theta values. This is the equation of a straight line:
    y_hat = intercept + slope * X.

    error = predictions - y: The difference between the model's predictions and the
    actual y values is calculated. This is how far off the model is.

    gradient_theta0 = np.sum(error) / m: This calculates the gradient (the
    direction and magnitude of the steepest ascent) for the intercept (theta[0]).
    It's the average of the errors.

    gradient_theta1 = np.sum(error * X) / m: This calculates the gradient for the
    slope (theta[1]). It's the average of the product of errors and the X values.

    theta[0] = theta[0] - alpha * gradient_theta0: The intercept (theta[0]) is
    updated by moving it in the opposite direction of its gradient, scaled by the
    learning rate alpha. This minimizes the error.

    theta[1] = theta[1] - alpha * gradient_theta1: Similarly, the slope (theta[1])
    is updated using its gradient and the learning rate.
    """

    # Initialize some useful variables
    m = X.shape[0]

    # Loop over the number of iterations
    for _ in range(num_iters):
        # Calculate the predictions for the current theta values
        predictions = theta[0] + theta[1] * X

        # Calculate the error (difference between predictions and actual values)
        error = predictions - y

        # Calculate the gradient for theta[0] (intercept)
        gradient_theta0 = np.sum(error) / m

        # Calculate the gradient for theta[1] (slope)
        gradient_theta1 = np.sum(error * X) / m

        # Update theta parameters using the learning rate (alpha)
        theta[0] = theta[0] - alpha * gradient_theta0
        theta[1] = theta[1] - alpha * gradient_theta1

    return theta
```

```python
theta = np.zeros(2)
visualize(theta, X, y)
theta = fit(X, y, theta, 0.01, 1)
visualize(theta, X, y)
theta = fit(X, y, theta, 0.01, 2)
visualize(theta, X, y)
theta = fit(X, y, theta, 0.01, 3)
visualize(theta, X, y)
theta = fit(X, y, theta, 0.01, 4)
visualize(theta, X, y)
theta = fit(X, y, theta, 0.01, 1000)
visualize(theta, X, y)
```

Predicting salary for 10 and 15 years (modifying the iterations, the result varies):

```python
theta = np.zeros(2)
theta = fit(X, y, theta, 0.01, 150)
predict_10 = predict(10, theta)
predict_15 = predict(15, theta)
print(f"Predicted salary for 10 years of experience: {predict_10}")
print(f"Predicted salary for 15 years of experience: {predict_15}")
```
