Getting the Car's Companys:

```python
df['CarCompany'] = df['CarName'].str.split().str[0]
```

checking the unique values under "CarCompany:

```python
df['CarCompany'].unique()
```

Fixing wrong names:

```python
df['CarCompany']=df['CarCompany'].replace('vw', 'volkswagen').replace('vokswagen', 'volkswagen')
df['CarCompany']=df['CarCompany'].replace('maxda', 'mazda')
df['CarCompany']=df['CarCompany'].replace('Nissan', 'nissan')
df['CarCompany']=df['CarCompany'].replace('toyouta', 'toyota')
df['CarCompany']=df['CarCompany'].replace('porcshce', 'porsche')
df['CarCompany']=df['CarCompany'].replace('alfa-romero', 'alfa-romeo')
```
