Getting the Car's Companys:

```python
df['CarCompany'] = df['CarName'].str.split().str[0]
```

checking the unique values under "CarCompany:

```python
df['CarCompany'].unique()
```

Fixing wrong names:

```python
df['CarCompany']=df['CarCompany'].replace(['vw', 'vokswagen'], 'volkswagen')
df['CarCompany']=df['CarCompany'].replace('maxda', 'mazda')
df['CarCompany']=df['CarCompany'].replace('Nissan', 'nissan')
df['CarCompany']=df['CarCompany'].replace('toyouta', 'toyota')
df['CarCompany']=df['CarCompany'].replace('porcshce', 'porsche')
df['CarCompany']=df['CarCompany'].replace('alfa-romero', 'alfa-romeo')
```

```python
df['CarCompany'].unique()
```

Creating a Dict with CarCompany and price data:

```python
company_price = dict(df.groupby('CarCompany')['price'].mean())
company_price = {key: float(value) for key, value in company_price.items()}
company_price
```

```python
company_price = df.groupby('CarCompany')['price'].mean()
my_array = pd.cut(company_price, bins=[0, 10000, 20000, float('inf')], labels=["low", "med"
, "high"], right=False)
```

```python
df['CarCompany'] = df['CarCompany'].map(my_array)
df['CarCompany'].unique()
```

```python
df['CarCompany_high'] = (df['CarCompany'] == 'high').astype(int)
df['CarCompany_low'] = (df['CarCompany'] == 'low').astype(int)
df['CarCompany_med'] = (df['CarCompany'] == 'med').astype(int)
df = df.drop(columns="CarCompany")
df.head()
```
